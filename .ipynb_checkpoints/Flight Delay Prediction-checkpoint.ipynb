### Project Description

The Airlines dataset comprises 539,383 instances and encompasses 8 distinct features. The objective is to predict whether a given flight will experience a delay, based on the scheduled departure information.

### Data Description

| Column         | Description                                              |
|----------------|----------------------------------------------------------|
| Airline        | Different types of commercial airlines                   |
| Flight         | Types of Aircraft                                        |
| AirportFrom    | Source Airport                                           |
| AirportTo      | Destination Airport                                      |
| DayOfWeek      | Indicates the day of the week                            |
| Time           | Length of Delay                                          |
| DepartureTime  | Departure time measured in minutes from midnight (range: 10-1439) |
| Duration       | Duration of the flight in minutes                        |
| Delay          | Whether the flight is delayed or not                     |

Considering the presence of right skewness in these feature***s, the Yeo-Johnson transformat***ion is a suitable choice for addressing skewness. The Yeo-Johnson transformation is capable of handling both positive and negative skewness, making it versatile for various types of skewed distributions. By applying the Yeo-Johnson transformation to the 'Flight' and 'Length' features, we aim to normalize their distributions, making them more symmetrical and aligning them better with the assumptions of machine learning algorithms. This preprocessing step could potentially enhance the performance of predictive models by reducing bias and improving the stability of parameter estimates.

Therefore, the decision to use the Yeo-Johnson transformation is justified based on the observed skewness in the dataset and the desire to prepare the data appropriately for subsequent modelingasks.

***Inference and Justification***<br>
Based on the original class distribution, the imbalance ratio is approximately 0.80, indicating that there is a moderate imbalance between the two classes. In this case, the majority class (Delay=0) has significantly more samples than the minority class (Delay=1).

Given the imbalance ratio exceeding the threshold of 0.3, it suggests that data balancing techniques may be beneficial to address the class imbalance. Therefore, implementing techniques like SMOTE (Synthetic Minority Over-sampling Technique) to oversample the minority class or undersample the majority class could help improve model performance and mitigate the effects of class imbalance.

Applying SMOTE to oversample the minority class can help in several ways: Improved Predictive Performance, Reduced Bias and Preservation of Information

Therefore, in the context of predicting flight delays, where accurate identification of delayed flights is crucial, applying SMOTE is justified. It enables the development of predictive models that are more robust, accurate, and capable of handling the inherent class imbalance in the dataset, ultimately improving decision-making processes in aviation and related industries.


```python
#SMOTE to  balance the classes in the training data

# Separate the target variable (Delay) and features
X = dataset_clean.drop(columns=['Delay'])
y = dataset_clean['Delay']

# Encode 'DayOfWeek' separately to ensure it has seven columns representing each day of the week
X_encoded = pd.get_dummies(X, columns=['DayOfWeek'])

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Get the list of non-numeric columns before applying SMOTE
non_numeric_columns = X_train.select_dtypes(exclude=['float64', 'int64']).columns

# Split the categorical and numeric features
X_train_numeric = X_train.drop(columns=non_numeric_columns)

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_numeric, y_train)

# Check the balance of classes after SMOTE
print("Class distribution after SMOTE:")
print(pd.Series(y_train_resampled).value_counts())

# Now, encode the remaining categorical features using one-hot encoding
remaining_categorical_cols = list(set(non_numeric_columns) - set(['DayOfWeek'])) # Exclude 'DayOfWeek'
X_train_resampled_categorical = X_train[remaining_categorical_cols].reindex(X_train_resampled.index)
X_train_resampled_encoded = pd.get_dummies(X_train_resampled_categorical, columns=remaining_categorical_cols)

# Concatenate the resampled numeric features with the encoded categorical features
X_train_resampled_final = pd.concat([X_train_resampled, X_train_resampled_encoded], axis=1)

# Check the shape of the training data after encoding
print("Shape of X_train_resampled after encoding:", X_train_resampled_final.shape)
```

```python
# Split the resampled dataset into training and test sets Train = 80 % Test = 20%
X_train, X_test, y_train, y_test = train_test_split(X_train_resampled_final, y_train_resampled, test_size=0.2, random_state=42)
```

```python
# Build logistic regression model
logistic_regression_model = LogisticRegression(random_state=42)
logistic_regression_model.fit(X_train, y_train)

# Build decision tree model
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)
```

### Model Evaluation:

- **Accuracy**: Refers to the overall percentage of correctly classified instances. In this case, the Decision Tree model has a higher accuracy than the Logistic Regression model.
- **Precision**: Describes the ratio of true positive predictions to the total positive predictions. The Decision Tree has a slightly better precision than Logistic Regression.
- **Recall** (or Sensitivity): Indicates the ratio of true positive predictions to the total actual positives. Logistic Regression has a perfect recall, meaning it correctly identified all actual positives but also had a high false positive rate.
- **F1-score**: Is the harmonic mean of precision and recall, giving a balance between the two in cases where they may trade off. The Decision Tree has a higher F1-score, suggesting a better balance between precision and recall.
- **ROC-AUC**: The area under the receiver operating characteristic curve, a plot of the true positive rate against the false positive rate. A higher ROC-AUC value generally indicates a better model, but in this case, both models have similar ROC-AUC values, with the Decision Tree being slightly better.
